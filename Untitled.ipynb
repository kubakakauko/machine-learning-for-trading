```python
%matplotlib widget
import matplotlib.pyplot as plt
import ccxt
import pandas as pd
import numpy as np
import talib as ta
from platypus import Problem, Real, NSGAII
import backtrader as bt
import time
from IPython.display import Image

# Download crypto data from Binance
def download_crypto_data(symbol, timeframe='1d', limit=1000):
    exchange = ccxt.binance()
    since = exchange.parse8601('2021-01-01T00:00:00Z')
    data = []
    
    while since < exchange.milliseconds():
        try:
            ohlcv = exchange.fetch_ohlcv(symbol, timeframe=timeframe, since=since, limit=limit)
            if len(ohlcv) == 0:
                break
            data += ohlcv
            since = ohlcv[-1][0] + 1
            time.sleep(exchange.rateLimit / 1000)
        except Exception as e:
            print(f"Error fetching data: {e}")
            break
    
    df = pd.DataFrame(data, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('timestamp', inplace=True)
    return df

# Calculate alpha factors
def calculate_alpha_factors(data):
    alphas = pd.DataFrame(index=data.index)
    
    def rank(series):
        return series.rank(pct=True)

    def ts_rank(series, window):
        return series.rolling(window).apply(lambda x: rank(x).iloc[-1])

    def ts_argmax(series, window):
        # Ensure the input is a pandas Series
        series = pd.Series(series)
        return series.rolling(window).apply(lambda x: np.argmax(x)) + 1

    def correlation(x, y, window):
        return x.rolling(window).corr(y)

    def alpha_1(data):
        returns = data['close'].pct_change()
        cond = np.where(returns < 0, ta.STDDEV(returns, timeperiod=20), data['close'])
        cond_series = pd.Series(cond, index=data.index)
        return rank(ts_argmax(np.power(cond_series, 2), 5)) - 0.5

    def alpha_2(data):
        volume_log_diff = rank(np.log(data['volume']).diff(2))
        close_open_ratio = rank((data['close'] - data['open']) / data['open'])
        return -1 * correlation(volume_log_diff, close_open_ratio, 6)

    def alpha_3(data):
        open_rank = rank(data['open'])
        volume_rank = rank(data['volume'])
        return -1 * correlation(open_rank, volume_rank, 10)

    def alpha_4(data):
        return -1 * ts_rank(rank(data['low']), 9)

    def alpha_5(data):
        vwap = (data['close'] * data['volume']).rolling(10).sum() / data['volume'].rolling(10).sum()
        return rank(data['open'] - vwap) * (-1 * abs(rank(data['close'] - vwap)))

    def alpha_6(data):
        return -1 * correlation(data['open'], data['volume'], 10)

    alphas['alpha_1'] = alpha_1(data)
    alphas['alpha_2'] = alpha_2(data)
    alphas['alpha_3'] = alpha_3(data)
    alphas['alpha_4'] = alpha_4(data)
    alphas['alpha_5'] = alpha_5(data)
    alphas['alpha_6'] = alpha_6(data)
    
    return alphas

# Define the optimization problem
class AlphaOptimization(Problem):
    def __init__(self, data, num_alphas=6):
        super(AlphaOptimization, self).__init__(num_alphas, 2)  # 2 objectives: Sharpe and Volatility
        self.data = data
        self.types[:] = [Real(0, 1) for _ in range(num_alphas)]  # Weights for each alpha

    def evaluate(self, solution):
        weights = solution.variables
        portfolio_returns = sum(weights[i] * self.data[f'alpha_{i+1}'] for i in range(len(weights)))
        sharpe_ratio = portfolio_returns.mean() / portfolio_returns.std()
        volatility = portfolio_returns.std()
        solution.objectives[:] = [-sharpe_ratio, volatility]  # Minimize volatility, maximize Sharpe

# Backtest Strategy
class AlphaStrategy(bt.Strategy):
    def __init__(self, alphas):
        self.alphas = alphas
        self.data0 = self.datas[0]  # Assuming only one data feed

    def next(self):
        weights = self.alphas
        alpha_combination = sum(weights[i] * self.data0.lines[i] for i in range(len(weights)))
        if alpha_combination > 0:
            self.buy()
        elif alpha_combination < 0:
            self.sell()

# Main workflow
if __name__ == "__main__":
    # Download data
    btc_data = download_crypto_data('BTC/USDT')

    # Calculate alpha factors
    alpha_data = calculate_alpha_factors(btc_data)

    # Define and run optimization
    problem = AlphaOptimization(alpha_data)
    algorithm = NSGAII(problem)
    algorithm.run(100000)

    # Retrieve results
    best_solution = min(algorithm.result, key=lambda s: s.objectives[0])  # Maximize Sharpe Ratio
    print("Best Weights:", best_solution.variables)
    
     # Backtesting
    cerebro = bt.Cerebro()
    cerebro.addstrategy(AlphaStrategy, alphas=best_solution.variables)
    
    # Add data
    data = bt.feeds.PandasData(dataname=btc_data)
    cerebro.adddata(data)
    
    # Run the backtest
    cerebro.run()
    
    # Save plot to file
    figs = cerebro.plot(style='candlestick', barup='green', bardown='red')
    for i, fig in enumerate(figs):
        fig[0].savefig(f'backtrader_plot_{i}.png')  # Save each figure with a unique name
    
    # Optionally display the saved plot
    for i in range(len(figs)):
        display(Image(filename=f'backtrader_plot_{i}.png'))
```
